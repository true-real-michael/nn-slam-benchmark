```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open('/Users/mikhailkiselyov/code/thesis/nn-slam-benchmark/measurements/orange/st_lucia_feature_lightglueb.pkl', 'rb') as f:
    data = pickle.load(f)

data = {
    int(k[12:]): 1549 /v['feature_matching'] for k, v in data.items()
}
```

```python
data
```

```python
x = [str(k) for k in data]
y = [data[k] for k in data]

fig, ax = plt.subplots()
ax.bar(x, y)
for bar in ax.patches:
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, round(bar.get_height(), 2), ha='center', va='bottom')

plt.title('LightGlue throughput')
plt.xlabel('Number of features')
plt.ylabel('Throughput, FPS')
plt.show()
```
